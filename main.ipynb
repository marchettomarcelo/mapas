```python
import geopandas as gpd
import matplotlib.pyplot as plt

infos = gpd.read_file('mapa.gpkg', layer='lim_unidade_federacao_a')
```

```python
# lista de latitudes e longitudes
lat = [-25]
long = [-50]

# lista de latitudes e longitudes das fábricas de embalagens
lat_embalagens = [-23, -24]
long_embalagens = [-48, -50]

infos = infos.to_crs(epsg=4326)
ax = infos.plot(color='white', edgecolor='black', figsize=(15, 8))

# plotar pontos (aqui você pode plotar os pontos de grupos diferentes com cores diferentes)
ax.scatter(long, lat, s=100, c='blue', marker='.')
ax.scatter(long_embalagens, lat_embalagens, s=100, c='red', marker='.')
plt.title('Fábricas da Klabin no Brasil')
plt.show()


```
